```python

from faker import Faker
f=Faker(locale='zh_CN')
#VEHICLE to case
import csv
import numpy as np
import random
#生成车牌号
def chepaihao():
  char0='京津沪渝冀豫云辽黑湘皖鲁新苏浙赣鄂桂甘晋蒙陕吉闽赣粤青藏川宁琼'
  char1='ABCDEFGHJKLMNPQRSTUVWXYZ'#车牌号中没有I和O，可自行百度
  char2='1234567890'
  len0=len(char0)-1
  len1 = len(char1) - 1
  len2 = len(char2) - 1
  code = ''
  for i in range(0,1):
    index0 = random.randint(1,len0 )
    index1 = random.randint(1, len1)
    code += char0[index0]
    code += char1[index1]
    for i in range(1, 6):
      index2 = random.randint(1, len2)
      code += char2[index2]
    #print(code,end=',')
  return code  

if __name__=='__main__':
    list = []
    for i in range(0,200):
       a=(
           #entity 1:
           #case
           f.random_int(min=2001,max=2999),#case ID
           f.random_int(min=20000001,max=29999999),
           f.date_time(),
           f.date_time(),
           f.date_time(),
           f.date_time(),
           f.date_time(),
           f.pyfloat(left_digits=4,right_digits=2 ,positive=True),
           f.company_prefix(),
           f.random_int(min=30000001,max=39999999),
           f.random_digit(),#0-9的整数
           f.phone_number(),
           f.phone_number(),
           f.date_time(),
           f.address(),
           f.name(),
           "备注",
           '无',
           f.date_time(),
           
           #entity 2
           #vehicle
           f.random_int(min=70001,max=79999),#vehicle ID
           f.random_int(min=70000001,max=79999999),#regist NO
           f.name(),
           f.random_element(),
           f.ssn(),#驾驶员编号？
           f.pystr(),
           f.pystr(),
           chepaihao(),
           chepaihao(),#三者车牌号？
           f.pystr(),
           f.pystr(),#kind
           f.date_time(),#ETL_DATE
           f.name(),#LAST_MARKER
           "备注",
           '无',
           f.date_time(),
           
           #relationships
           f.random.choice(['Accident_Vehicle_OF_CASE','Third_Party_Vehicle_OF_CASE']),
           f.pystr(),
           f.pystr(),
          
         )
       list.append(a)
    b=np.array(list)
    print(b[0])
    with open('vehicle to case.csv', 'w', encoding='utf-8', newline='') as ff:
        writer = csv.writer(ff)
        writer.writerows(b)        
```
