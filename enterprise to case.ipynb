```python

from faker import Faker
f=Faker(locale='zh_CN')
#enterprise to case
import csv
import numpy as np
import random
list = []
for i in range(0,200):
       a=(
           #entity 1:
           #enterprise
           f.random_int(min=1001,max=1999),#index 15
           f.random_int(min=10000001,max=19999999),
           f.company_suffix(),#公司性质
           f.company_prefix(),#公司名称
           f.random_int(),
           f.pystr(),
           f.random.choice(['大','中','小']),
           f.phone_number(),
           f.pystr(),
           f.date_time(),#ETL_DATE
           f.name(),#LAST_MARKER
           "备注",#MARKED_STATUS
           '无',#MARKED_TAG
           f.date_time(),
           
           #entity 2
           #case
           f.random_int(min=2001,max=2999),#index 15
           f.random_int(min=20000001,max=29999999),
           f.date_time(),
           f.date_time(),
           f.date_time(),
           f.date_time(),
           f.date_time(),
           f.pyfloat(left_digits=4,right_digits=2 ,positive=True),
           f.company_prefix(),
           f.random_int(min=30000001,max=39999999),
           f.random_digit(),#0-9的整数
           f.phone_number(),
           f.phone_number(),
           f.date_time(),
           f.address(),
           f.name(),
           "备注",
           '无',
           f.date_time(),
           
           #relationships
           f.random.choice(['ASSESSMENT']),
           f.pystr(),
           f.pystr(),
          
         )
       list.append(a)
b=np.array(list)
print(b[0])
with open('enterprise to case.csv', 'w', encoding='utf-8', newline='') as ff:
    writer = csv.writer(ff)
    writer.writerows(b)        
```
